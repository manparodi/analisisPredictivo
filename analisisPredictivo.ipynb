# Normalizacion de los datos



```python
#Copio el dataframe para no modificar mi dataframe original y selecciono solo datos numericos
df1 = df.select_dtypes(include ="float")
print(df1)
```

```python
#Uso la estandarizacion robusta debido a la alta cantidad de valores atipicos
from sklearn.preprocessing import RobustScaler
transformer = RobustScaler() 
np_norm = transformer.fit_transform(df1)
df1 = pd.DataFrame(np_norm, columns=df1.columns)
df1
```

```python
df1.describe()
```

# Grafico de Perfiles

```python
import plotly.graph_objects as go

import pandas as pd


fig = go.Figure(data=
    go.Parcoords(
        line = dict(
                   color = 'black'),
        dimensions = list([
            dict(range = [-3,17],
                constraintrange = [4,8],
                label = 'platelets', values = df1['platelets']),
             dict(range = [-4,17],
                label = 'serum_sodium', values = df1['serum_sodium']),    
            dict(range = [-4,17],
                label = 'age', values = df1['age']),
            dict(range = [-4,17],
                label = 'creatinine_phosphokinase', values = df1['creatinine_phosphokinase']),
            dict(range = [-4,17],
                label = 'time', values = df1['time']),
            dict(range = [-4,17],
                label = 'serum_creatinine', values = df1['serum_creatinine']),
               dict(range = [-4,17],
                label = 'ejection_fraction', values = df1['ejection_fraction']),
        ])
    )
)
fig.show()
```

```python
fig = go.Figure(data=
    go.Parcoords(
        line = dict(
                   colorscale = [[0,'purple'],[0.5,'lightseagreen'],[1,'gold']]),
        dimensions = list([
              dict(range = [-4,3],
                constraintrange = [4,8],
                label = 'ejection_fraction', values = df1['ejection_fraction']),
            dict(range = [-4,3],
                label = 'age', values = df1['age']),
            dict(range = [-4,3],
                label = 'serum_sodium', values = df1['serum_sodium']),
              dict(range = [-4,3],
                label = 'time', values = df1['time'])
        ])
    )
)

fig.update_layout(
    plot_bgcolor = 'white',
    paper_bgcolor = 'white'
)

fig.show()
```
